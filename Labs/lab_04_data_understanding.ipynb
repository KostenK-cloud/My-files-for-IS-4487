We will be using a San Francisco rental unit dataset.  Some rentals are apartments, others are for homes, and there may be some other random properties for rent.  Each row represents one rental listing.

**Dataset:** `rent.csv`

## Part 3: Basic Exploration

Use `groupby` and `value_counts` to summarize trends across neighborhoods and cities.  The concept of data exploration will be covered more in week 5.


```python
# Average price by neighborhood
df.groupby('nhood')['price'].mean().sort_values(ascending=False).head(10)
```

```python
# Top cities by count
df['city'].value_counts().head(10)
```
